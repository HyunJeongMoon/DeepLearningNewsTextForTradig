  - > kill pid번호
* TimeZone 맞춘다
  - > GCP의 locale이 아이오와, -9시간


```python
import schedule
import time
from datetime import datetime, timedelta
import subprocess
import pytz

def run_sokbo():
    print('start run_sokbo')
    # subprocess 모듈을 사용하여 py_today_sokbo_process.py를 실행
    subprocess.run(["python", "py_today_sokbo_process.py"])

# 서울의 시간을 얻는 함수
def get_seoul_time():
    seoul_timezone = pytz.timezone('Asia/Seoul')
    seoul_time = datetime.now(seoul_timezone)
    return seoul_time

# 서울 시간을 기준으로 23:58과 08:59에 run_test 함수를 실행하도록 스케줄링
def schedule_jobs():
    seoul_time = get_seoul_time()

    print (datetime.now(), 'now...')
    
    # 23시 58분에 실행
    schedule_time_1 = seoul_time.replace(hour=23, minute=58, second=0, microsecond=0)- timedelta(hours=9)
    schedule.every().day.at(schedule_time_1.strftime("%H:%M")).do(run_sokbo)
    print("1", schedule_time_1)

    # 08시 59분에 실행
    schedule_time_2 = seoul_time.replace(hour=8, minute=59, second=0, microsecond=0) - timedelta(hours=9)
    schedule.every().day.at(schedule_time_2.strftime("%H:%M")).do(run_sokbo)
    print("2", schedule_time_2)

schedule_jobs()

while True:
    # 스케줄링된 작업 실행
    schedule.run_pending()
    # 실행확인용 화면 출력
    print (get_seoul_time())
    # 시간 대기
    time.sleep(1)
```
